```python
import random
def generate_random_array(size):
    return [random.randint(0, 100) for _ in range(size)]
```

```python
MAX = float("inf")
MIN = float("-inf")

def minimax_test(depth, scores, index, maxTurn, alpha, beta, target_depth):
    if depth == target_depth:
        return scores[index]
    
    if maxTurn:
        best = MIN
        for i in range(0,3):
            value = minimax_test(depth+1, scores, index*3+i, False, alpha, beta, target_depth)
            best = max(best, value)
            alpha  = max(best, alpha)
            if alpha >= beta:
                break
            return best
    else:
        best = MAX
        for i in range(0,3):
            value = minimax_test(depth+1, scores, index*3+i, True, alpha, beta, target_depth)
            best = min(best, value)
            beta  = min(best, beta)
            if alpha >= beta:
                break
            return best    
        

if __name__ == '__main__':
    given_scores = generate_random_array(27)
    
    print("OPTIMAL VALUE IS : ", end="")
    print(minimax_test(0,given_scores, 0, False, MIN, MAX, 3))        
```

```python
def minimax_again(values, index, maxTurn, depth, target_depth=3):
    if depth == target_depth:
        return values[index]
    
    if maxTurn:
        return max (
            minimax_again(values, index * 3 + 0, False, depth+1, target_depth),
            minimax_again(values, index * 3 + 1, False, depth+1, target_depth),
            minimax_again(values, index * 3 + 2, False, depth+1, target_depth)
        )
    else:
        return min (
            minimax_again(values, index * 3 + 0, True, depth+1, target_depth),
            minimax_again(values, index * 3 + 1, True, depth+1, target_depth),
            minimax_again(values, index * 3 + 2, True, depth+1, target_depth)
        )    
        
if __name__ == '__main__':
    given_scores = generate_random_array(27)
    print(given_scores)
    print("OPTIMAL VALUE IS : ", end="")
    print(minimax_test(0,given_scores, 0, True, MIN, MAX, 3))            
```

```python
class WaterJugProblem:
    def __init__(self):
        self.visited = set() 
        self.solutions = []  

    def is_goal_state(self, state):
        return state[0] == 2

    def get_neighbors(self, state):
        x, y = state
        neighbors = []
        if x < 4:
            neighbors.append((4, y))
        if y < 3:
            neighbors.append((x, 3))
        if x > 0:
            neighbors.append((0, y))
        if y > 0:
            neighbors.append((x, 0))
        if x > 0 and y < 3:
            transfer = min(x, 3 - y)
            neighbors.append((x - transfer, y + transfer))
        if y > 0 and x < 4:
            transfer = min(y, 4 - x)
            neighbors.append((x + transfer, y - transfer))

        return neighbors

    def dfs(self, state, path):
        if self.is_goal_state(state):
            self.solutions.append(path + [state]) 
            return True

        if state in self.visited:
            return False

        self.visited.add(state)

        for neighbor in self.get_neighbors(state):
            if self.dfs(neighbor, path + [state]):
                return True

        return False

    def solve(self):
        initial_state = (0, 0) 
        self.dfs(initial_state, [])
        return self.solutions

if __name__ == "__main__":
    water_jug_problem = WaterJugProblem()
    solutions = water_jug_problem.solve()
    
    if solutions:
        print("Solution found:")
        for solution in solutions:
            print("Path to solution:", solution)
    else:
        print("No solution found.")
```

```python
class WaterJugProblemShort:
    def __init__(self):
        self.visited = set([])
        self.solutions = []

    def is_goal_state(self, state):
        return state[0] == 2

    def get_neighbors(self, state):
        x, y = state
        neighbors = []
        if x < 4:
            neighbors.append((4, y))
        if y < 3:
            neighbors.append((x, 3))
        if x > 0:
            neighbors.append((0, y))
        if y > 0:
            neighbors.append((x, 0))
        if x > 0 and y < 3:
            transfer = min(x, 3 - y)
            neighbors.append((x - transfer, y + transfer))
        if y > 0 and x < 4:
            transfer = min(y, 4 - x)
            neighbors.append((x + transfer, y - transfer))

        return neighbors

    def dfs(self, state, path):
        if self.is_goal_state(state):
            self.solutions.append(path + [state])
        elif state not in self.visited:
            self.visited.add(state)
            for neighbor in self.get_neighbors(state):
                self.dfs(neighbor, path + [state])

    def solve(self):
        self.dfs((0, 0), [])
        return self.solutions

if __name__ == "__main__":
    water_jug_problem = WaterJugProblemShort()
    solutions = water_jug_problem.solve()
    if solutions:
        print("Solution found:")
        for solution in solutions:
            print("Path to solution:", solution)
    else:
        print("No solution found.")
```

```python
# Define the goal state
GOAL_STATE = [[1, 2, 3], [5, 8, 6], [0, 7, 4]]

# Calculate the Manhattan distance
def manhattan_distance(state):
    distance = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] != 0:
                goal_x, goal_y = (state[i][j] - 1) // 3, (state[i][j] - 1) % 3
                distance += abs(goal_x - i) + abs(goal_y - j)
    return distance

# Find the blank tile (0) position
def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

# Generate all possible moves
def generate_moves(state):
    moves = []
    x, y = find_blank(state)
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # Up, Down, Left, Right
    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = [row[:] for row in state]
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            moves.append(new_state)
    return moves

# Find the state with the smallest f-cost
def get_next_state(open_list):
    best_state = min(open_list, key=lambda x: x[0])
    open_list.remove(best_state)
    return best_state

# A* Search
def a_star(initial_state):
    open_list = [(manhattan_distance(initial_state), 0, initial_state, [])]  # (f, g, state, path)
    visited = []

    while open_list:
        f, g, current, path = get_next_state(open_list)
        if current == GOAL_STATE:
            return path + [current]

        if current not in visited:
            visited.append(current)
            for move in generate_moves(current):
                open_list.append((g + 1 + manhattan_distance(move), g + 1, move, path + [current]))

    return None

# Test the implementation
initial_state = [[1, 2, 3], [5, 6, 0], [7, 8, 4]]
solution = a_star(initial_state)

if solution:
    print("Solution steps:")
    for step in solution:
        for row in step:
            print(row)
        print()
else:
    print("No solution found.")

```
